## The duration feature is highly correlated with the target variable y, but it is not a predictor in the true sense of the term. The key reason for this is:

- Correlation vs. Predictive Value: While the duration feature shows a high correlation with the target variable y, this correlation does not necessarily imply causation or predictive power. In this case, the duration feature reflects the length of time a customer is contacted or engaged. However, the presence of this feature often aligns with the fact that customers are only contacted when there are existing issues or problems.

- Contextual Interpretation: The duration of a call or interaction does not inherently predict the likelihood of a customer having issues or being successful. Instead, it serves as an outcome of the underlying process where customers are reached out to primarily when problems are identified. Thus, the duration itself becomes an outcome rather than a cause or predictor of the customerâs status or behavior.

- Predictive Models: In predictive modeling, itâs crucial to distinguish between features that truly contribute to the prediction of the target variable and those that are correlated due to their outcome-related nature. Features like duration that are closely tied to the target outcome but do not contribute additional predictive value can be misleading. Including such features might inflate correlation metrics but wonât improve the modelâs true predictive performance.

- Conclusion: The duration feature, while correlated with the target variable, should be carefully evaluated in the context of its role in the predictive model. It is more a result of the model's decision-making process rather than a feature that provides predictive power. Therefore, it should be treated accordingly in feature selection and model evaluation processes.

# Performance curves
- The ROC curve helps to visualize the trade-off between the True Positive Rate and False Positive Rate.


```python
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, roc_curve, auc, precision_recall_curve

# Get the predicted probabilities for the test set
y_prob = model_xg.predict_proba(X_test_transformed)[:, 1]  # Probability estimates for the positive class

# Compute ROC curve
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

# Precision-Recall Curve

## Bagging (Bootstrap Aggregating):
- Bagging is an ensemble technique where multiple instances of a model are trained on different bootstrap samples of the training data.
- Each model's predictions are aggregated (e.g., by averaging or majority vote) to form the final prediction.
- Bagging helps reduce variance and avoids overfitting by combining the results of individual models.
- Random Forest is a classic example of a bagging method, where multiple decision trees are trained on different subsets of data and their predictions are averaged.

## Boosting:
- Boosting is an ensemble technique that builds models sequentially, where each model attempts to correct the errors of its predecessor.
- Boosting focuses more on the samples that were previously misclassified, giving them higher weights in subsequent models.
- This approach reduces bias and can significantly improve model accuracy.
- XGBoost is a popular boosting method that builds a strong predictive model by iteratively adding models to correct errors made by previous models.
